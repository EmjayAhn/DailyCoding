# 단일 Thread
- Main Thread 하나만 사용 하였을 때 결과

```python
from selenium import webdriver
import time
from threading import Thread
page_numbers = 40
timer = {}
```

```python
total_article_dict = {}

def crawling_alone():
    start_time = time.time()
    urls = []
    for i in range(page_numbers):
        urls.append("https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=105#&date=%2000:00:00&page={}".format(i+1))
    
    driver = webdriver.Chrome()
    for url in urls:
        driver.get(url)
        article_list = []
        time.sleep(1)
        driver.execute_script("window.scrollTo(0, 2500)")
        section_body = driver.find_element_by_css_selector(".section_body")
        targets = section_body.find_elements_by_css_selector("li dt:nth-child(2)")
        for target in targets:
            article_list.append(target.text)
        try:
            int(url[-2])
            total_article_dict[url[-2]+url[-1]+' page'] = article_list
        except:
            total_article_dict[url[-1]+' page'] = article_list
    driver.quit()
    end_time = time.time() - start_time
    timer['SingleThread'] = round(end_time, 4)
    
def uni_thread():
    t = Thread(target=crawling_alone)
    t.start()
```

```python
uni_thread()
```

# Dual Thread
- MainThread 안에 2개의 SubThread로 Crawling
- 2개의 SubTread 는 각각 짝수 페이지와 홀수 페이지를 돌린다.

```python
total_article_dict = {}

def crawling_odd():
    start_time = time.time()
    urls = []
    for i in range(page_numbers // 2):
        urls.append("https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=105#&date=%2000:00:00&page={}".format(2*i + 1))
    
    article_list = []
    driver = webdriver.Chrome()
    for url in urls:
        driver.get(url)
        time.sleep(1)
        driver.execute_script("window.scrollTo(0, 2500)")
        section_body = driver.find_element_by_css_selector(".section_body")
        targets = section_body.find_elements_by_css_selector("li dt:nth-child(2)")
        for target in targets:
            article_list.append(target.text)
        try:
            int(url[-2])
            total_article_dict[url[-2]+url[-1]+' page'] = article_list
        except:
            total_article_dict[url[-1]+' page'] = article_list
    driver.quit()
    end_time = time.time() - start_time
    timer['DualThread'] = round(end_time, 4)
        
    
def crawling_even():
    start_time = time.time()
    urls = []
    for i in range(page_numbers // 2):
        urls.append("https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=105#&date=%2000:00:00&page={}".format(2*i+2))
    
    article_list = []
    driver = webdriver.Chrome()
    for url in urls:
        driver.get(url)
        time.sleep(1)
        driver.execute_script("window.scrollTo(0, 2500)")
        section_body = driver.find_element_by_css_selector(".section_body")
        targets = section_body.find_elements_by_css_selector("li dt:nth-child(2)")
        for target in targets:
            article_list.append(target.text)
        try:
            int(url[-2])
            total_article_dict[url[-2]+url[-1]+' page'] = article_list
        except:
            total_article_dict[url[-1]+' page'] = article_list
    driver.quit()
    end_time = time.time() - start_time
    timer['DualThread'] = round(end_time, 4)

def dual_threads():
    t1 = Thread(target=crawling_odd)
    t2 = Thread(target=crawling_even)
    t1.start()
    t2.start()
```

# Quad Thread
- 4개의 Thread 로 돌려봤을때,
- 각각 1~10, 11~20, 21~30, 31~40의 페이지를 크롤링 해온다

```python
total_article_dict = {}

def crawling(start, end):
    start_time = time.time()
    urls = []
    for i in range(start, end+1):
        urls.append("https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=105#&date=%2000:00:00&page={}".format(i))
    
    article_list = []
    driver = webdriver.Chrome()
    for url in urls:
        driver.get(url)
        time.sleep(1)
        driver.execute_script("window.scrollTo(0, 2500)")
        section_body = driver.find_element_by_css_selector(".section_body")
        targets = section_body.find_elements_by_css_selector("li dt:nth-child(2)")
        for target in targets:
            article_list.append(target.text)
        try:
            int(url[-2])
            total_article_dict[url[-2]+url[-1]+' page'] = article_list
        except:
            total_article_dict[url[-1]+' page'] = article_list
    driver.quit()
    end_time = time.time() - start_time
    timer['Quad Thread'] = end_time

def my_threads():
    t1 = Thread(target=crawling, args=(1, 10))
    t2 = Thread(target=crawling, args=(11, 20))
    t3 = Thread(target=crawling, args=(21, 30))
    t4 = Thread(target=crawling, args=(31, 40))
    t1.start()
    t2.start()
    t3.start()
    t4.start()
```

```python
my_threads()
```

# Octa Thread
- 8개의 Thread 로 돌려봤을 떄
- 각각 5개씩 돌린다.

```python
total_article_dict = {}

def crawling(start, end):
    start_time = time.time()
    urls = []
    for i in range(start, end+1):
        urls.append("https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=105#&date=%2000:00:00&page={}".format(i))
    
    article_list = []
    driver = webdriver.Chrome()
    for url in urls:
        driver.get(url)
        time.sleep(1)
        driver.execute_script("window.scrollTo(0, 2500)")
        section_body = driver.find_element_by_css_selector(".section_body")
        targets = section_body.find_elements_by_css_selector("li dt:nth-child(2)")
        for target in targets:
            article_list.append(target.text)
        try:
            int(url[-2])
            total_article_dict[url[-2]+url[-1]+' page'] = article_list
        except:
            total_article_dict[url[-1]+' page'] = article_list
    driver.quit()
    end_time = time.time() - start_time
    timer['Octa Thread'] = end_time
    
def my_threads():
    threads = []
    for i in range(8):
        threads.append(Thread(target=crawling, args=(5*i+1, 5*i+5)))
    for thread in threads:
        thread.start()
```
